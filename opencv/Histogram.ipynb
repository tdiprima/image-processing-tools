```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('incept4.png')
```

```python
# Extract the green channel
green_channel = image[:, :, 1]  # BGR
```

```python
# Mean
np.mean(green_channel)
```

```python
# Standard Deviation
np.std(green_channel)
```

```python
# Compute histogram
hist, bins = np.histogram(green_channel.flatten(), bins='auto', range=[0, 256])
```

```python
# Plot histogram
plt.figure()
plt.bar(bins[:-1], hist, width=bins[1] - bins[0], align='center', color='green')
plt.title("Probability Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()

# Doesn't look like it's really sure of anything:
```

```python
# Plot histogram
plt.figure()
plt.bar(bins[:-1], hist, width=bins[1] - bins[0], align='center', color='green')
plt.title("Probability Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.ylim(0, 40000)
plt.xlim(0, 50)
plt.show()

# Modified axes so we can see a little better:
```

```python
# Create Pandas DataFrame from Image Channels

from PIL import Image
import pandas as pd
import numpy as np

# Open the image file
img = Image.open('incept4.png')

# Convert the image to an array
img_array = np.array(img)

# Get the R, G, B channels
R = img_array[:,:,0]
G = img_array[:,:,1]
B = img_array[:,:,2]

# Flatten the arrays and create a DataFrame
df = pd.DataFrame({
    'R': R.flatten(),
    'G': G.flatten(),
    'B': B.flatten()
})
```

```python
plt.imshow(img)
```

```python
df.describe()
```

```python
df
```

```python
from matplotlib import pyplot as plt
df['G'].plot(kind='hist', color='green', bins=20, title='G')
plt.gca().spines[['top', 'right',]].set_visible(False)

# PROBABILITY:
```

```python
from matplotlib import pyplot as plt
df['R'].plot(kind='hist', color='red', bins=20, title='R')
plt.gca().spines[['top', 'right',]].set_visible(False)

# R ZERO IS BACKGROUND:
```

```python
from matplotlib import pyplot as plt
df['G'].plot(kind='line', color='green', figsize=(8, 4), title='G')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
from matplotlib import pyplot as plt
df['R'].plot(kind='line', color='red', figsize=(8, 4), title='R')
plt.gca().spines[['top', 'right']].set_visible(False)

# A LOT OF ZEROES, AND AN OUTLIER
```

```python
# NEXT! "BY CLASS"
image1 = cv2.imread('class.png')

# Extract the red channel
red_channel = image1[:, :, 2]  # BGR

plt.imshow(Image.open('class.png'))
```

```python
hist1, bins1 = np.histogram(red_channel.flatten(), bins='auto', range=[0, 256])

# Plot histogram
plt.figure()
bar_width = bins1[1] - bins1[0]
plt.bar(bins1[:-1], hist1, width=bar_width, align='center', color='red')
plt.title("Class Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.xlim(0, 150)
plt.show()

# 128 IS BACKGROUND.  OTHER THAN THAT, WE CAN SEE THAT IT'S MOSTLY ONE CLASS:
```
